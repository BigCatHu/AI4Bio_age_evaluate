```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('/mnt/Bioage/filled_data.csv')
df = df.iloc[1:, :]


df.dtypes
```

```python

df2=df.astype(float)
```

```python
df_nan_corr = df2.corr()
print(df_nan_corr)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df_nan_corr, vmax=1, vmin=-1, center=0)
plt.savefig('seaborn_heatmap_corr_example.png')

```

```python
# 生成浅色调色板
light_blue = sns.light_palette("steelblue", as_cmap=True)
light_pink = sns.light_palette("lightpink", as_cmap=True)
# light_orange = sns.light_palette("lightorange", as_cmap=True)
# sns.clustermap(df_nan_corr,    #仅仅需传入绘图数据集
#               )
# 绘制热力图，并设置颜色映射为浅色调色板
sns.heatmap(df_nan_corr,
            cmap=light_blue,  # 设置为浅蓝色
            vmax=1,
            vmin=-1,
            center=0,
            annot=True,
            fmt=".2f",
            linewidths=0.5,
            annot_kws={"size": 8},
            cbar_kws={"shrink": 0.8})

# 调整坐标轴标签的字体大小
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# 保存图像
# plt.savefig('seaborn_heatmap_corr_example.png', dpi=300, bbox_inches='tight')
```

```python
import seaborn as sns

# 绘制分层聚类分析图
cluster_map = sns.clustermap(df_nan_corr, annot=True)

# 保存图像
cluster_map.savefig('cluster_map.png',dpi = 300)

# 显示图形
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 假设 df2 是包含要绘制的特征的数据列

sns.displot(data=df2, bins=20, kde=True, rug=True)
plt.title('Feature distribution')
plt.savefig('feature_distribution.png', dpi = 300)  # 将图像保存为名为 feature_distribution.png 的文件
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 假设 df2 是包含要绘制的特征的数据列

# 设置颜色主题
sns.set_palette("Set3")

# 绘制分布图
sns.displot(data=df2, bins=20, kde=True, rug=True, color='skyblue', line_kws={'linestyle': '-'})
plt.title('Feature distribution')
# plt.savefig('feature_distribution.png', dpi=300)  # 将图像保存为名为 feature_distribution.png 的文件
plt.show()
```
